```python
# merge tables
merge1 = skus.merge(clicks)
merge2 = merge1.merge(users)
merge3 = merge2.merge(orders)
```

```python
# drop columns
merge = merge3.drop(['sku_ID', 'brand_ID', 'activate_date', 'deactivate_date', 'user_ID', 'request_time', 
                     'first_order_month', 'order_ID', 'order_date', 'order_time'], axis = 1)
```

```python
# take a look at the merged table
merge.info()
```

```python
# feature scaling
from sklearn.preprocessing import OrdinalEncoder
ord_enc = OrdinalEncoder()
merge['attribute1'] = ord_enc.fit_transform(merge[['attribute1']])
merge['attribute2'] = ord_enc.fit_transform(merge[['attribute2']])
merge['channel'] = ord_enc.fit_transform(merge[['channel']])
merge['gender'] = ord_enc.fit_transform(merge[['gender']])
merge['age'] = ord_enc.fit_transform(merge[['age']])
merge['marital_status'] = ord_enc.fit_transform(merge[['marital_status']])
merge['promise'] = ord_enc.fit_transform(merge[['promise']])
```

```python
merge.head()
```

```python
# reorder columns
def movecol(df, cols_to_move = [], ref_col = '', place = 'After'):
    
    cols = df.columns.tolist()
    if place == 'After':
        seg1 = cols[:list(cols).index(ref_col) + 1]
        seg2 = cols_to_move
    if place == 'Before':
        seg1 = cols[:list(cols).index(ref_col)]
        seg2 = cols_to_move + [ref_col]
    
    seg1 = [i for i in seg1 if i not in seg2]
    seg3 = [i for i in cols if i not in seg1 + seg2]
    
    return(df[seg1 + seg2 + seg3])
```

```python
merge = movecol(merge, cols_to_move=['purchase_power'], ref_col = 'dc_des', place = 'After')
merge.head()
```

```python
# set independent dependent variables
X = merge.iloc[:, :-1].values
y = merge.iloc[:, -1].values
```

```python
# check independent variables
print(X)
```

```python
# check dependent variable
print(y)
```

```python
# split into training and test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
# multiple linear regression
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
# compare y_pred and y_test
y_pred = regressor.predict(X_test)
np.set_printoptions(precision = 2)
print(np.concatenate((y_pred.reshape(len(y_pred), 1), y_test.reshape(len(y_test), 1)), 1))
```

```python
# mse and r2
print(mean_squared_error(y_test, y_pred))
print(r2_score(y_test, y_pred))
```

```python
# decision tree regression
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(random_state = 0)
regressor.fit(X, y)
```

```python
# y_pred
y_pred = regressor.predict(X_test)
```

```python
# mse and r2
print(mean_squared_error(y_test, y_pred))
print(r2_score(y_test, y_pred))
```

```python
# random forest regression
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators = 10, random_state = 0)
regressor.fit(X, y)
```

```python
# y_pred
y_pred = regressor.predict(X_test)
```

```python
# mse and r2
print(mean_squared_error(y_test, y_pred))
print(r2_score(y_test, y_pred))
```
